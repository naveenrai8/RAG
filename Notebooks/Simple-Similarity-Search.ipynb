# Simple Similarity based Retrieval

```python
import os
from dotenv import load_dotenv
from langchain_openai import AzureChatOpenAI, AzureOpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter, RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores.utils import DistanceStrategy
from langchain_community.document_loaders import PyMuPDFLoader
```

```python
is_load = load_dotenv()
print(is_load)

endpoint = os.getenv("AZURE_ENDPOINT")
model = os.getenv("AZURE_OPENAI_EMBEDDING_LARGE_MODEL")

subscription_key = os.getenv("AZURE_OPENAI_API_KEY")
api_version = os.getenv("AZURE_OPENAI_API_VERSION")

openai_embedding_llm = AzureOpenAIEmbeddings( 
        api_version=api_version,
        azure_deployment=model,
   )
print(openai_embedding_llm)
openai_embedding_llm.embed_query("India")
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20,
    length_function=len,
    is_separator_regex=False,
)
```

```python
from pymupdf import pymupdf
import os
from langchain_core.documents import Document

pdf_loader = PyMuPDFLoader('/Users/naveenrai/Learning/On_Github/RAG/Data/HTTP-Status-Codes.pdf')
docs = pdf_loader.load()
```

```python
print(docs[0])
```

```python
texts = splitter.split_documents(docs)
print(len(texts))
for t in texts:
    print(t, end='\n\n')
```

```python
vs_faiss = FAISS.from_documents(documents=texts, embedding=openai_embedding_llm)
```

```python
similar_docs = vs_faiss.similarity_search_with_score("error code 404", k = 3)
for sd, score in similar_docs:
    print(sd.page_content)
    print(score)
    print()
```

```python
similar_docs = vs_faiss.similarity_search("What is AI", k = 2)
for sd in similar_docs:
    print(sd.page_content)
    print()
```

```python
similar_docs = vs_faiss.similarity_search_with_score("What is AI", k = 2)
for sd, score in similar_docs:
    print(sd.page_content)
    print(score)
    print()
```
